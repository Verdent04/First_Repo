### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
print(df.info())
print()
print(df.describe())
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing = df.isnull()
missing_count = df.isna().sum()
print(missing_count) # prints number of rows that have missing value from each column

missing_percent = df.isna().mean() * 100 
print ('\nPercentage of missing values per column: ')
print (missing_percent) #prints percentage of the column's missing value

missing_col = df.columns[df.isnull().any()]
print ('\nMissing columns: ,\n', missing_col)

print (missing.shape)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
new_df = df.isnull()
plt.figure(figsize=(10,6))
sns.heatmap(new_df)
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
print (df['RemoteWork'].isnull().sum())
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
column_impute = 'EdLevel'
most_frequent = df[column_impute].mode()[0] 
print('Most frequent value in Employment column:\n',most_frequent)

#Experiment on finding the column with the most missing values
#column_with_most_missing = missing.sum().idxmax() 
#most_miss_count = missing_count.max() 

#print (f"Column with most missing values: '{column_with_most_missing}' ({most_miss_count} missing)")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df[column_impute].fillna(most_frequent, inplace=True)
print (f"\nMissing values in '{column_impute}' after imputation: {df[column_impute].isnull().sum()}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
plot_data = df[column_impute]
plt.figure(figsize=(10,6))
sns.heatmap(plot_data) 
plt.show()
```
